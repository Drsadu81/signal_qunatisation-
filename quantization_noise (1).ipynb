### Hyperparameter $n $ for the range search on log scale

```python
# Let find the range of uniform distribution between 0.001 to 1 which represent scale bounds

r = -4 * np.random.rand()
n = 10**r

n_min = np.log10(0.0001)
print("The lower bound is : " ,n_min)

n_mxm = np.log10(1)
print("The upper bound is : " ,n_mxm)

print("val of n : " , n  ," in this range")
```

```python
M = 128  # length of signal

# Create a signal 
x = np.sin(4*np.pi/M * np.arange(M))
print("Length of x is :", len(x))

print(" signal x :",x)

# quantize signal 
a = np.round(3 * x)
xQ = 1/3 * a

print('\n')
print(" quantization of signal xQ : ",xQ)

e = xQ - x
print("error e : ", xQ - x)

print(a.min())
print(a.max())



# plot (quantized) signals
fig, ax1 = plt.subplots(figsize=(10, 4))
ax2 = ax1.twinx()
ax1.plot(x, 'r')
ax1.plot(xQ, 'b')
ax1.plot(e, 'g')
```
